```python
data = []
with open('data.txt', 'r') as file:
    for line in file:
        data.append(line.strip())
```

```python
# 각 리스트 초기화
gan_loss = []
val_mse = []
psnr = []
mse = []

for line in data:
    b = line.split()
    gan_loss.append(float(b[4].rstrip(';')))  # gan_loss 추가 ( ; 제거)

    # val_mse 값 추출 부분 수정
    if len(b) > 7 and '=' in b[7]:
        val_mse_str = b[7].rstrip(';').split('=')[-1]
        if val_mse_str:  # 빈 문자열이 아닌지 확인
            val_mse.append(float(val_mse_str))   # val_mse 추가
        else:
            print("val_mse_str is empty!")
    else:
        print("Error: Unexpected format in line:", line)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(gan_loss, label='GAN Loss', color='blue')
ax1.set_title('GAN Loss over Epochs')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(val_mse, label='Validation MSE', color='red')
ax2.set_title('Validation MSE over Epochs')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('MSE')
ax2.legend()

plt.tight_layout()
plt.show()
```

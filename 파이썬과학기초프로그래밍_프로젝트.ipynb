```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 읽기
data = []
with open('data.txt', 'r') as file:
    for line in file:
        data.append(line.strip())  # 각 줄에서 공백 제거 후 리스트에 추가

# 각 자료를 저장할 리스트 초기화
gan_loss = []  # GAN 손실
val_mse = []
psnr = []      # PSNR
mse = []

# 데이터를 처리하여 각 리스트에 값 추가
for line in data:
    line = line.replace(';', '')  # 세미콜론 제거
    b = line.split()  # 단어별로 분리

    # gan_loss 값 추출 및 추가
    gan_loss_value = float(b[b.index('gan_loss') + 2])
    gan_loss.append(gan_loss_value)

    # val_mse 값 추출 및 추가
    val_mse_value = float(b[b.index('val_mse') + 2])
    val_mse.append(val_mse_value)

    # PSNR 값 추출 및 추가
    psnr_value = float(b[b.index('PSNR') + 2])
    psnr.append(psnr_value)

    # MSE 값 추출 및 추가
    mse_value = float(b[b.index('MSE') + 2])
    mse.append(mse_value)

# NumPy 배열로 변환
gan_loss_array = np.array(gan_loss)
val_mse_array = np.array(val_mse)
psnr_array = np.array(psnr)
mse_array = np.array(mse)

# 통계 값 출력
print("GAN Loss - 평균: {}, 최대값: {}, 최소값: {}".format(np.mean(gan_loss_array), np.max(gan_loss_array), np.min(gan_loss_array)))
print("Validation MSE - 평균: {}, 최대값: {}, 최소값: {}".format(np.mean(val_mse_array), np.max(val_mse_array), np.min(val_mse_array)))
print("PSNR - 평균: {}, 최대값: {}, 최소값: {}".format(np.mean(psnr_array), np.max(psnr_array), np.min(psnr_array)))
print("MSE - 평균: {}, 최대값: {}, 최소값: {}".format(np.mean(mse_array), np.max(mse_array), np.min(mse_array)))

# 데이터 시각화
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# 각 그래프에 데이터 추가
ax1.plot(gan_loss_array, label='GAN Loss', color='blue')
ax1.set_title('GAN Loss over Epochs')  # 그래프 제목
ax1.set_xlabel('Epochs')  # X축 라벨
ax1.set_ylabel('Loss')  # Y축 라벨
ax1.legend()

ax2.plot(val_mse_array, label='Validation MSE', color='red')
ax2.set_title('Validation MSE over Epochs')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('MSE')
ax2.legend()

plt.tight_layout()
plt.show()

# PSNR과 MSE의 subplot
fig, (ax3, ax4) = plt.subplots(2, 1, figsize=(12, 10))

# PSNR 그래프
ax3.plot(psnr_array, label='PSNR', color='green')
ax3.set_title('PSNR over Epochs')
ax3.set_xlabel('Epochs')
ax3.set_ylabel('PSNR')
ax3.legend()

# MSE 그래프
ax4.plot(mse_array, label='Mean Squared Error', color='purple')
ax4.set_title('MSE over Epochs')
ax4.set_xlabel('Epochs')
ax4.set_ylabel('MSE')
ax4.legend()

plt.tight_layout()
plt.show()
```
